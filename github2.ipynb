```python
import pandas as pd
df= pd.read_excel("customer.xlsx")
df.head()
```

```python
df.info()
```

```python
df=df.dropna(subset=["CustomerID"])
df=df[df["Quantity"]>0]
```

```python
df["TotalPrice"]=df["Quantity"]*df["UnitPrice"]
```

```python
customer_revenue=df.groupby("CustomerID")["TotalPrice"].sum()
customer_revenue.sort_values(ascending=False).head(10)
```

```python
customer_orders=df.groupby("CustomerID")["InvoiceNo"].nunique()
customer_orders.sort_values(ascending=False).head(10)
```

```python
customer_revenue.to_excel("customer_analysis.xlsx")
```
